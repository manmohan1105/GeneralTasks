```python
# Import the nescessary packages
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd 
import pandas_bokeh
import seaborn as sns

from bokeh.plotting import figure, output_file, show
pandas_bokeh.output_notebook()
```

```python
# Loading the csv file to dataframe
df = pd.read_csv('data.csv')
df.head()
```

```python
# Creating a new dataframe from old dataframe by droping unnecessary coloumns

df2 = df.drop(['aclGroups', 'aclSource', 'aclUsers', 'acls',
       'authlistExpire', 'basebandSerialNumber', 'bmuUniqueId', 
        'buildVariant',  'builds', 'chipId',
        'created',  'deviceId', 'ecid', 'enabled',
       'euicceid', 'firstSeen',  'fusingStatus', 
       'identifiers',  'modifiedBy',
       'online', 'primateSerialNumber', 'productType', 'rapEcid', 'savageUid',
       'seen', 'seid',  'sessionAclGroups', 'sessionAclUsers',
       'sessionAcls', 'sessionDefaultInactivityPeriod',
       'sessionDefaultInactivityPeriodSource', 'sessionDefaultLifetime',
       'sessionId', 'sessionMaxInactivityPeriod',
       'sessionMaxInactivityPeriodSource', 'sessionMaxLifetime',
       'sessionTimeouts', 'sessionTimeoutsSourceDefaultLifetimeSource',
       'sessionTimeoutsSourceMaxLifetimeSource',  'stateReason',
       'summary', 'tag.', 'tags', 'udid', 'updated', 'usbHubId', 'usbHubPort',
       'yonkersEcid'], axis=1)
df2.head()
```

```python
# Finding the no. of connected and no. of disconected devices

Connected_disconnected_count= df2["connected"].value_counts()

Connected_disconnected_count.plot(kind='pie',
                                  figsize=(15, 6),
                                  autopct='%1.1f%%', 
                                  startangle=90,    
                                  shadow=True,       
                                  labels=None,         
                                  pctdistance=1.12,    
                            )

plt.title('Percentage of each connected devices type')
plt.axis('equal') # Sets the pie chart to look like a circle.
plt.legend(Connected_disconnected_count.index, loc='upper left')
plt.show()
```

```python
# Percentages of each producttype

product= df["productType"].value_counts()

product.plot(               kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,        
                            pctdistance=1.12,    
                            )

plt.title('Percentage of each product type')
plt.axis('equal')
plt.legend(product.index, loc='upper left')
plt.show()
```
